##Exploring Rent Velocity in New York City; Third-Order Derivatives

```python
nyc = dfh.drop(['Metro', 'CountyName', 'Neighborhood'], axis=1)[dfh.City.isin(['New York'])]
nyc = nyc.drop(['City', 'State'], axis=1).transpose()
nyc.columns = nyc.iloc[0]
nyc = nyc.drop('RegionName')

d1 = nyc-nyc.shift()
d2 = d1-d1.shift()
d3 = d2-d2.shift()
```

###Notes - Making sense of the third derivative in terms of time series data, rent:

x is rent
first derivative is change in rent 
second derivative is rate of change in rent (rent acceleration)
third derivative is the (rent jerk; change in acceleration)

from point a to point b:
a 0 change in rent indicates no increase or decrease in rent
a + change in rent indicates an increase in average rent from the month before
a - change in rent indicates a decrease in average rent from the month before

a 0 acceleration indicates a constant rate of change in rent
a + acceleration indicates a greater change in rent than the month before
a - acceleration indicates a smaller change in rent than the month before
average acceleration is the average rate of change in rent over time

a 0 jerk indicates a no change in acceleration; a constant rate at which rent changes
a + jerk indicates an increase in acceleration; an increase in the rate at which rent changes
a - jerk indicates a decrease in acceleration; a decrease in the rate at which rent changes


- the greater the jerk value, the more aggressive the change in pace
- +/- jerk indicates when rent is accelerating/decelerating
- the value of jerk indicates the magnitude of change, where 0 jerk indicates a constant acceleration, and thereby a constant rate of change in rent


I would use the third derivative of rent to identify periods in time where rent can be expected to increase or decrease.  When jerk is closer to 0, the current trend in increasing rent is predictable; rent increases at a constant rate.  Comparing plots of rent and its third derivative, jerk, shows that spikes in jerk precede sharp changes in rent.  The greater the value of jerk, the more volatile rent is expected to be.  The lesser the jerk, the smoother the trend in rent; as jerk dips, changes in rent become more subtle.

Instead of using rent in regression analysis, perhaps using jerk/ f'''(x) might yield more acceptable results.  In the case of taking an average jerk per neighborhood, areas with higher jerk values tend to be more volatile in rent, whereas lower jerk values can be associated with areas where trends in rent are less chaotic.

Also, a positive jerk value may indicate consistent fluctiation within a certain range, whereas a negative jerk indicates a larger range between min and max

Theoretically, a neighborhood that is on pace to increase in property value, or neighborhood value overall, is idenftified by a 0-jerk (low rent sensitivity), a close-to-zero acceleration (constant rate of change in rent), and a high velocity (constant increases in rent).  All three of these indicators point to a growth trend in a neighborhood.

##Inserting additional variables: Average Rent, Property Value, Zillow Home Value Index

```python
zri_vel = pd.DataFrame(d1.mean()).sort(columns=0, ascending=True).reset_index().rename(columns={0:'Rent Velocity', 'RegionName':'Neighborhood'})
zri_accl = pd.DataFrame(d2.mean()).sort(columns=0, ascending=True).reset_index().rename(columns={0:'Rent Acceleration', 'RegionName':'Neighborhood'})
zri_jerk = pd.DataFrame(d3.mean()).sort(columns=0, ascending=True).reset_index().rename(columns={0:'Rent Jerk', 'RegionName':'Neighborhood'})
zri_mean = pd.DataFrame(nyc.ix[-1]).reset_index().rename(columns={'RegionName':'Neighborhood', '2015-01':'Current Rent'})
```

```python
zhvi = pd.DataFrame(pd.read_csv(zhvi_all_homes))
zhvi_all = zhvi[zhvi['City'].isin(['New York'])][['RegionName','Zhvi']].rename(columns={'RegionName':'Neighborhood', 'Zhvi':'Current ZHVI'})
```

```python
df = zri_vel.merge(zri_mean, on='Neighborhood')
df = df.merge(zri_accl, on='Neighborhood')
df = df.merge(zri_jerk, on='Neighborhood')
df = df.merge(pd.DataFrame(zpsf.mean()).reset_index().rename(columns={0:'Average ZRI, $/SqFt', 'RegionName':'Neighborhood'}), on='Neighborhood')
df = df.merge(zhvi_all, on='Neighborhood')

df['Current Rent'] = [float(x) for x in df['Current Rent']]

df = df[['Neighborhood', 'Current Rent', 'Rent Velocity', 'Rent Acceleration', 'Rent Jerk', 'Average ZRI, $/SqFt', 'Current ZHVI']]
```

##Finalizing a Data Frame for NYC

```python
drop = [x for x in df[df['Neighborhood'].duplicated() == True].index] #dropping duplicated neighborhoods
nyc = df.drop(df.index[drop])

nyc['Current ZHVI'] = nyc['Current ZHVI']/100000
nyc['Current Rent'] = nyc['Current Rent']/1000
nyc = nyc.rename(columns={'Current Rent':'ZRI, $ (thousands)',
                          'Rent Velocity' : 'ZRI Velocity, $/month',
                          'Rent Acceleration' : 'ZRI Acceleration, $/month' + u'\xb2',
                          'Rent Jerk' : 'ZRI Jerk, $/month'  + u'\xb3',
                          'Average ZRI, $/SqFt':'ZRI, $/SqFt',
                          'Current ZHVI':'ZHVI, $ (ten-thousands)'})
```
